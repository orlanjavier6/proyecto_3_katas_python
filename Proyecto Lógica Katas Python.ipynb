1. Escribe una función que reciba una cadena de texto como parámetro y devuelva un diccionario con las frecuencias de cada letra en la cadena. Los espacios no deben ser considerados.

```python
def contar_frecuencias(cadena):
    #Creamos un diccionario vacío para almacenar los resultados
    frecuencias = {}
    
    #Recorremos cada caracter de la cadena
    for letra in cadena:
        #Verificamos que no sea un espacio 
        if letra != " ":
            #Pasamos a minúsculas para que 'A' y 'a' se cuenten como la misma letra
            letra = letra.lower()
            
            #Si la letra ya está en el diccionario, sumamos 1
            if letra in frecuencias:
                frecuencias[letra] += 1
            #Si no está, la inicializamos en 1
            else:
                frecuencias[letra] = 1
                
    return frecuencias

#Ejemplo de uso:
texto = "Hola Mundo"
resultado = contar_frecuencias(texto)
print(f"Frecuencias de '{texto}':", resultado)
```

5. Ecribe una función que tome una lista de números como parámetro y un valor opcional nota_aprobado, que por defecto es 5. La función debe calcular la media de los números en la lista y determinar si la media es mayor o igual que nota aprobado. Si es así, el estado será "aprobado", de lo contrario, será "suspenso". La función debe devolver
una tupla que contenga la media y el estado.


```python
def evaluar_media(lista_notas, nota_aprobado=5):
    #Calculamos la media: suma de todos los elementos entre la cantidad de ellos
    if not lista_notas: #Caso de seguridad por si la lista está vacía
        return (0, "suspenso")
        
    media = sum(lista_notas) / len(lista_notas)
    
    #Determinamos el estado según la nota de corte
    if media >= nota_aprobado:
        estado = "aprobado"
    else:
        estado = "suspenso"
        
    #Devolvemos la tupla requerida
    return (media, estado)

#Casos de uso
print(evaluar_media([4, 6, 8]))               #Usa el valor por defecto (5) -> (6.0, 'aprobado')
print(evaluar_media([4, 6, 8], nota_aprobado=7)) #Usa un valor personalizado -> (6.0, 'suspenso')
```

6. Escribe una función que calcule el factorial de un número de manera recursiva.

```python
def factorial(n):
    #Caso base: El factorial de 0 o 1 es siempre 1
    if n == 0 or n == 1:
        return 1
    #Caso recursivo: n! = n * (n-1)!
    else:
        return n * factorial(n - 1)

#Ejemplo de uso:
numero = 5
resultado = factorial(numero)
print(f"El factorial de {numero} es: {resultado}")
#Salida: 120 (porque 5 * 4 * 3 * 2 * 1 = 120)
```

7. Genera una función que convierta una lista de tuplas a una lista de strings. Usa la función map()

```python
def convertir_a_strings(lista_tuplas):
    #Usamos map para juntar los elementos de cada tupla en un solo string. El join sirve para unir lo que hay dentro de la tupla con un espacio
    resultado = list(map(lambda t: " ".join(map(str, t)), lista_tuplas))
    
    return resultado

#Para probar que funciona:
datos = [("Juan", 25), ("Ana", 30), ("Python", 3.10)]
print(convertir_a_strings(datos))
#Salida: ['Juan 25', 'Ana 30', 'Python 3.1']
```

```python
class ErrorListaVacia(Exception):
    """Excepción lanzada cuando la lista de números no tiene elementos."""
    pass

def calcular_promedio(numeros):
    try:
        #Comprobamos si la lista está vacía [cite: 19]
        if len(numeros) == 0:
            raise ErrorListaVacia("No se puede calcular el promedio de una lista vacía.")
        
        #Si tiene números, calculamos el promedio [cite: 18]
        promedio = sum(numeros) / len(numeros)
        return promedio
        
    except ErrorListaVacia as e:
        #Manejamos el error adecuadamente [cite: 19]
        print(f"Error detectado: {e}")
        return None

#Ejemplos de uso:
print(f"Promedio de [10, 20, 30]: {calcular_promedio([10, 20, 30])}")
print(f"Promedio de []: {calcular_promedio([])}")
```

12. Genera una función que al recibir una frase devuelva una lista con la longitud de cada palabra. Usa la función map()

```python
def medir_palabras(frase):
    #Primero dividimos la frase en una lista de palabras usando split()
    palabras = frase.split()
    
    #Usamos map() para aplicar la función len (longitud) a cada palabra, len nos dirá cuántos caracteres tiene cada elemento
    resultado = list(map(len, palabras))
    
    return resultado

#Ejemplo para probarlo:
texto_ejemplo = "Python es genial para el master"
longitudes = medir_palabras(texto_ejemplo)

print(f"Frase: '{texto_ejemplo}'")
print(f"Longitudes: {longitudes}")
#Salida: [6, 2, 6, 4, 2, 6]
```

13. Genera una función la cual, para un conjunto de caracteres, devuelva una lista de tuplas con cada letra en mayúsculas y minúsculas. Las letras no pueden estar repetidas .Usa la función map()


```python
def transformar_caracteres(texto):
    #Primero, usamos set() para eliminar automáticamente cualquier letra repetida. Esto cumple con el requisito de "Las letras no pueden estar repetidas"
    caracteres_unicos = set(texto.replace(" ", ""))
    
    #Usamos map() para convertir cada letra en una tupla (MAYÚSCULA, minúscula)
    resultado = list(map(lambda letra: (letra.upper(), letra.lower()), caracteres_unicos))
    
    return resultado

#Ejemplo de uso:
entrada = "abracadabra"
print(f"Resultado para '{entrada}':")
print(transformar_caracteres(entrada))
```

```python
from functools import reduce

def lista_a_numero(digitos):
    #La función reduce combina los elementos de la lista de izquierda a derecha. Lógica: Para [5, 7, 2], primero hace (5 * 10 + 7) = 57, luego (57 * 10 + 2) = 572.
    numero = reduce(lambda acumulado, digito: acumulado * 10 + digito, digitos)
    
    return numero

#Ejemplo de uso:
mis_digitos = [5, 7, 2]
resultado = lista_a_numero(mis_digitos)

print(f"Lista original: {mis_digitos}")
print(f"Número resultante: {resultado}")
#Salida: 572
```

22. Dada una lista numérica, obtén el producto total de los valores de dicha lista. Usa la función reduce() .

```python
from functools import reduce

def calcular_producto_total(lista_numeros):
    # La función reduce toma los dos primeros elementos, los multiplica, y el resultado lo multiplica por el tercero, y así sucesivamente. Lógica: ((((1 * 2) * 3) * 4) ...)
    producto = reduce(lambda x, y: x * y, lista_numeros)
    
    return producto

#Ejemplo de uso:
numeros = [1, 2, 3, 4, 5]
resultado = calcular_producto_total(numeros)

print(f"Lista de números: {numeros}")
print(f"El producto total es: {resultado}")
#Salida: 120 (porque 1*2*3*4*5 = 120)
```

24. Calcula la diferencia total en los valores de una lista. Usa la función reduce() .

```python
from functools import reduce

def calcular_diferencia_total(lista_numeros):
    #La función reduce toma el primer elemento y le resta el segundo, luego al resultado le resta el tercero, y así sucesivamente. 
    #Lógica para [100, 20, 10]: (100 - 20) - 10 = 70
    diferencia = reduce(lambda acumulado, siguiente: acumulado - siguiente, lista_numeros)
    
    return diferencia

#Ejemplo de uso:
numeros = [100, 25, 15, 10]
resultado = calcular_diferencia_total(numeros)

print(f"Lista de números: {numeros}")
print(f"La diferencia total es: {resultado}")
#Salida: 50 (porque 100 - 25 - 15 - 10 = 50)
```

27. Crea una función que calcule el promedio de una lista de números.

```python
def calcular_promedio(lista_numeros):
    #Primero comprobamos que la lista no esté vacía para evitar errores
    if len(lista_numeros) == 0:
        return 0
    
    #El promedio es la suma de todos los elementos dividida por la cantidad total
    suma_total = sum(lista_numeros)
    cantidad = len(lista_numeros)
    
    return suma_total / cantidad

#Ejemplo de uso:
mis_notas = [8, 9, 10, 7, 9]
resultado = calcular_promedio(mis_notas)

print(f"El promedio de la lista {mis_notas} es: {resultado}")
#Salida: 8.6
```

28. Crea una función que busque y devuelva el primer elemento duplicado en una lista dada.

```python
def encontrar_primer_duplicado(lista):
    #Usamos un conjunto para guardar los elementos que ya hemos procesado. Los sets en Python son muy rápidos para buscar si algo ya existe
    vistos = set()
    
    for elemento in lista:
        if elemento in vistos:
            #Si el elemento ya estaba en 'vistos', hemos encontrado el primer duplicado
            return elemento
        #Si no estaba, lo añadimos al conjunto y seguimos
        vistos.add(elemento)
    
    #Si terminamos el bucle sin encontrar nada, devolvemos None
    return None

#Ejemplo de uso:
numeros = [1, 2, 3, 4, 2, 5, 3]
resultado = encontrar_primer_duplicado(numeros)

if resultado is not None:
    print(f"El primer elemento duplicado es: {resultado}")
else:
    print("No se encontraron duplicados.")
#Salida: El primer elemento duplicado es: 2
```

29. Crea una función que convierta una variable en una cadena de texto y enmascare todos los caracteres con el carácter '#', excepto los últimos cuatro.

```python
def enmascarar_variable(variable):
    #Primero, convertimos cualquier variable a string (cadena de texto)
    texto = str(variable)
    
    #Si el texto es muy corto (4 caracteres o menos), no hace falta enmascarar
    if len(texto) <= 4:
        return texto
    
    #Creamos la máscara: 1. Tomamos la longitud total menos 4 y la multiplicamos por '#'. 2. Le sumamos (concatenamos) los últimos 4 caracteres usando slicing [-4:]
    resultado = "#" * (len(texto) - 4) + texto[-4:]
    
    return resultado

#Ejemplos de uso:
print(f"Tarjeta: {enmascarar_variable('4548123456789012')}")
#Salida: ############9012

print(f"Teléfono: {enmascarar_variable(600123456)}")
#Salida: #####3456
```

30. Crea una función que determine si dos palabras son anagramas, es decir, si están formadas por las mismas letras pero en diferente orden.

```python
def son_anagramas(palabra1, palabra2):
    #Primero, normalizamos las palabras: 1. Las pasamos a minúsculas (.lower()). 2. Quitamos los espacios (.replace(" ", ""))
    p1 = str(palabra1).lower().replace(" ", "")
    p2 = str(palabra2).lower().replace(" ", "")
    
    #Si tienen diferente longitud tras limpiar espacios, es imposible que sean anagramas.
    if len(p1) != len(p2):
        return False
    
    #La clave: ordenamos las letras de ambas palabras, sorted() devuelve una lista con las letras en orden alfabético. Si las listas ordenadas son iguales, son anagramas.
    return sorted(p1) == sorted(p2)

#Ejemplos de uso:
print(f"¿'Roma' y 'Amor' son anagramas?: {son_anagramas('Roma', 'Amor')}")
print(f"¿'Frase' y 'Fresa' son anagramas?: {son_anagramas('Frase', 'Fresa')}")
print(f"¿'Hola' y 'Adios' son anagramas?: {son_anagramas('Hola', 'Adios')}")
```

36. Crea la clase UsuarioBanco ,representa a un usuario de un banco con su nombre, saldo y si tiene o no cuenta corriente. Proporciona métodos para realizar operaciones como retirar dinero, transferir dinero desde otro usuario y agregar dinero al saldo.

Código a seguir:
1.Inicializar un usuario con su nombre, saldo y si tiene o no cuenta corriente mediante True y False .
2.Implementar el método retirar_dinero para retirar dinero del saldo del usuario. Lanzará un error en caso de no poder hacerse.
3.Implementar el método transferir_dinero para realizar una transferencia desde otro usuario al usuario actual. Lanzará un error en caso de no poder hacerse.
4.Implementar el método agregar_dinero para agregar dinero al saldo del usuario.

Caso de uso:
1.Crear dos usuarios: "Alicia" con saldo inicial de 100 y "Bob" con saldo inicial de 50, ambos con cuenta corriente.
2.Agregar 20 unidades de saldo de "Bob".
3.Hacer una transferencia de 80 unidades desde "Bob" a "Alicia".
4.Retirar 50 unidades de saldo a "Alicia".

```python
#Definimos excepciones personalizadas para que el banco sea robusto
class SaldoInsuficienteError(Exception):
    """Excepción lanzada cuando no hay saldo suficiente para la operación."""
    pass

class UsuarioBanco:
    #1.Inicializar un usuario con su nombre, saldo y si tiene o no cuenta corriente (True/False)
    def __init__(self, nombre, saldo_inicial, cuenta_corriente):
        self.nombre = nombre
        self.saldo = float(saldo_inicial)
        self.cuenta_corriente = bool(cuenta_corriente)

    #2.Implementar el método retirar_dinero
    def retirar_dinero(self, cantidad):
        if cantidad <= 0:
            raise ValueError("La cantidad a retirar debe ser mayor que cero.")
        
        if self.saldo < cantidad:
            raise SaldoInsuficienteError(
                f"Operación denegada: {self.nombre} intenta retirar {cantidad}, "
                f"pero su saldo actual es de {self.saldo}."
            )
        
        self.saldo -= cantidad
        print(f"Retirada con éxito: {self.nombre} ha retirado {cantidad}. Nuevo saldo: {self.saldo}")

    #3.Implementar el método transferir_dinero (desde otro usuario al usuario actual)
    def transferir_dinero(self, emisor, cantidad):
        if cantidad <= 0:
            raise ValueError("La cantidad a transferir debe ser mayor que cero.")
        
        print(f"Iniciando transferencia de {cantidad} desde {emisor.nombre} hacia {self.nombre}...")
        
        #Primero intentamos retirar el dinero del emisor. Si el emisor no tiene saldo, este método lanzará SaldoInsuficienteError automáticamente.
        emisor.retirar_dinero(cantidad)
        
        #Si la retirada fue exitosa, agregamos el dinero al receptor (self)
        self.agregar_dinero(cantidad)
        print(f"¡Transferencia completada con éxito!")

    #4.Implementar el método agregar_dinero
    def agregar_dinero(self, cantidad):
        if cantidad <= 0:
            raise ValueError("La cantidad a agregar debe ser mayor que cero.")
        
        self.saldo += cantidad
        print(f"Ingreso con éxito: Se han añadido {cantidad} a la cuenta de {self.nombre}. Nuevo saldo: {self.saldo}")

    def info_usuario(self):
        return f"Usuario: {self.nombre} | Saldo: {self.saldo} | CC: {self.cuenta_corriente}"


#---EJECUCIÓN ESTRICTA DEL CASO DE USO---

print("=== SIMULACIÓN BANCARIA ===")

try:
    #1.Crear dos usuarios: "Alicia" con saldo inicial de 100 y "Bob" con saldo inicial de 50, ambos con cuenta corriente.
    alicia = UsuarioBanco("Alicia", 100, True)
    bob = UsuarioBanco("Bob", 50, True)
    
    print(alicia.info_usuario())
    print(bob.info_usuario())
    print("-" * 40)

    #2.Agregar 20 unidades de saldo a "Bob".
    bob.agregar_dinero(20)
    print("-" * 40)

    #3.Hacer una transferencia de 80 unidades desde "Bob" a "Alicia". Nota: Bob empezó con 50 y sumó 20, tiene 70. Transferir 80 lanzará un error.
    alicia.transferir_dinero(emisor=bob, cantidad=80)
    print("-" * 40)

except SaldoInsuficienteError as e:
    print(f"\n[ALERTA SISTEMA] {e}\n")
    print("-" * 40)

#4.Retirar 50 unidades de saldo a "Alicia". Ejecutamos este paso fuera del bloque anterior por si la transferencia falló (como es el caso), asegurando que Alicia pueda continuar con su operación.
try:
    alicia.retirar_dinero(50)
    print("-" * 40)
except SaldoInsuficienteError as e:
    print(f"\n[ALERTA SISTEMA] {e}\n")

print("\n=== ESTADO FINAL DE LAS CUENTAS ===")
print(alicia.info_usuario())
print(bob.info_usuario())
```

Código a seguir:
1. Crear una función contar_palabras para contar el número de veces que aparece cada palabra en el texto. Tiene que devolver un diccionario.
2. Crear una función reemplazar_palabras para remplazar una palabra_original del texto por una palabra_nueva . Tiene que devolver el texto con el remplazo de palabras.
3. Crear una función eliminar_palabra para eliminar una palabra del texto. Tiene que devolver el texto con la palabra eliminada.
4. Crear la función procesar_texto que tome un texto, una opción(entre "contar", "reemplazar", "eliminar") y un número de argumentos variable según la opción indicada.

Caso de uso:
Comprueba el funcionamiento completo de la función procesar_texto

```python
#1.Función para contar el número de veces que aparece cada palabra
def contar_palabras(texto):
    #Pasamos a minúsculas y limpiamos signos básicos para contar bien
    palabras = texto.lower().replace(",", "").replace(".", "").split()
    frecuencias = {}
    
    for palabra in palabras:
        frecuencias[palabra] = frecuencias.get(palabra, 0) + 1
        
    return frecuencias


#2.Función para reemplazar una palabra por otra
def reemplazar_palabras(texto, palabra_original, palabra_nueva):
    #Usamos replace directamente sobre la cadena de texto
    return texto.replace(palabra_original, palabra_nueva)


#3.Función para eliminar una palabra específica por completo
def eliminar_palabra(texto, palabra_a_eliminar):
    palabras = texto.split()
    #Filtramos la lista para excluir la palabra que queremos borrar
    palabras_filtradas = [p for p in palabras if p != palabra_a_eliminar]
    return " ".join(palabras_filtradas)


#4.Función principal que actúa como "orquestadora"
def procesar_texto(texto, opcion, *args):
    """
    Procesa un texto según la opción ('contar', 'reemplazar', 'eliminar').
    *args captura de forma dinámica los parámetros adicionales necesarios.
    """
    if opcion == "contar":
        return contar_palabras(texto)
        
    elif opcion == "reemplazar":
        #'reemplazar' necesita dos argumentos extra: palabra_original y palabra_nueva
        if len(args) < 2:
            raise ValueError("La opción 'reemplazar' requiere: palabra_original y palabra_nueva.")
        return reemplazar_palabras(texto, args[0], args[1])
        
    elif opcion == "eliminar":
        #'eliminar' necesita un argumento extra: palabra_a_eliminar
        if len(args) < 1:
            raise ValueError("La opción 'eliminar' requiere: palabra_a_eliminar.")
        return eliminar_palabra(texto, args[0])
        
    else:
        return "Error: Opción no válida. Elige entre 'contar', 'reemplazar' o 'eliminar'."


#---CASO DE USO (Comprobación del funcionamiento completo)---

texto_prueba = "Python es genial y aprender Python te abre muchas puertas"
print(f"Texto original: '{texto_prueba}'\n")

#Prueba 1: contar palabras
resultado_contar = procesar_texto(texto_prueba, "contar")
print("1. Resultado Opción 'contar':")
print(resultado_contar)
print("-" * 50)

#Prueba 2: reemplazar palabras. Pasamos "Python" y "JavaScript" como argumentos variables (*args)
resultado_reemplazar = procesar_texto(texto_prueba, "reemplazar", "Python", "JavaScript")
print("2. Resultado Opción 'reemplazar' (Python -> JavaScript):")
print(f"'{resultado_reemplazar}'")
print("-" * 50)

#Prueba 3: eliminar palabra. Pasamos "genial" como argumento variable (*args)
resultado_eliminar = procesar_texto(texto_prueba, "eliminar", "genial")
print("3. Resultado Opción 'eliminar' (palabra: 'genial'):")
print(f"'{resultado_eliminar}'")
print("-" * 50)
```

40. Escribe una función que tome dos parámetros: figura (una cadena que puede ser "rectangulo" , "circulo" o "triangulo" ) y datos (una tupla con los datos necesarios para calcular el área de la figura).

```python
import math

def calcular_area(figura, datos):
    """
    Calcula el área de una figura geométrica.
    - figura: "rectangulo", "circulo" o "triangulo"
    - datos: tupla con las medidas necesarias
    """
    #Normalizamos el nombre de la figura a minúsculas y sin espacios
    figura_limpia = figura.strip().lower()
    
    if figura_limpia == "rectangulo":
        #Esperamos (base, altura)
        if len(datos) != 2:
            return "Error: El rectángulo requiere 2 datos (base, altura)."
        base, altura = datos
        area = base * altura
        return f"El área del rectángulo es: {area}"
        
    elif figura_limpia == "triangulo":
        #Esperamos (base, altura)
        if len(datos) != 2:
            return "Error: El triángulo requiere 2 datos (base, altura)."
        base, altura = datos
        area = (base * altura) / 2
        return f"El área del triángulo es: {area}"
        
    elif figura_limpia == "circulo":
        #Esperamos (radio,) -> Nota: la coma es obligatoria en tuplas de un solo elemento
        if len(datos) != 1:
            return "Error: El círculo requiere 1 dato (radio)."
        radio = datos[0]
        area = math.pi * (radio ** 2)
        return f"El área del círculo es: {round(area, 2)}"
        
    else:
        return f"Error: La figura '{figura}' no está soportada. Usa 'rectangulo', 'circulo' o 'triangulo'."

#---CASOS DE PRUEBA---

print(calcular_area("rectangulo", (10, 5)))   #Salida: el área del rectángulo es: 50
print(calcular_area("triangulo", (8, 4)))     #Salida: el área del triángulo es: 16.0
print(calcular_area("circulo", (3,)))         #Salida: el área del círculo es: 28.27 (Aproximado)
print(calcular_area("pentagono", (5, 2)))     #Salida: mensaje de error controlado
```

```python
def calcular_total_tienda():
    print("=== PASARELA DE PAGO: TIENDA EN LÍNEA ===")
    try:
        #1.Solicita al usuario que ingrese el precio original de un artículo.
        precio_original = float(input("Introduce el precio original del artículo (€): "))
        
        if precio_original <= 0:
            print("Error: El precio del artículo debe ser un número positivo mayor que cero.")
            return

        #2.Pregunta al usuario si tiene un cupón de descuento (respuesta sí o no).
        tiene_cupon = input("¿Dispone de un cupón de descuento? (sí/no): ").strip().lower()

        #Inicializamos el descuento en cero por defecto
        descuento = 0.0

        #3.Si el usuario responde que sí, solicita que ingrese el valor del cupón.
        if tiene_cupon in ["sí", "si", "s"]:
            valor_cupon = float(input("Introduce el valor del cupón (€): "))
            
            #4.Aplica el descuento siempre y cuando el valor del cupón sea válido
            if valor_cupon > 0:
                if valor_cupon > precio_original:
                    #Validación extra lógica: El descuento no puede superar el precio del producto
                    print(f"Aviso: El cupón ({valor_cupon}€) supera el precio del producto. La compra será gratuita.")
                    descuento = precio_original
                else:
                    descuento = valor_cupon
                    print(f"¡Cupón aplicado con éxito! Descuento de {descuento}€.")
            else:
                print("Error: El valor del cupón no es válido (debe ser mayor a cero). No se aplicará descuento.")
        
        elif tiene_cupon in ["no", "n"]:
            print("No se ha aplicado ningún cupón de descuento.")
        
        else:
            print("Opción no reconocida. Se procesará la compra sin descuento.")

        #5.Muestra el precio final de la compra, teniendo en cuenta el descuento aplicado o sin él.
        precio_final = precio_original - descuento
        
        print("\n=== RESUMEN DE COMPRA ===")
        print(f"Precio original: {precio_original:.2f}€")
        print(f"Descuento aplicado: {descuento:.2f}€")
        print(f"MONTO FINAL A PAGAR: {precio_final:.2f}€")
        print("=========================")

    except ValueError:
        print("Error: Por favor, introduce un valor numérico válido para los precios y cupones.")

#Ejecutamos el programa para verificar su funcionamiento
calcular_total_tienda()
```
